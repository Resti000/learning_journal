```python
# break (untuk berhenti sesuai perintah)
for i in range(1,11):
  if i == 4 :
    break
  print(i)
```

```python
# continue (untuk skip perintah yg dituliskan)
for i in range(1,11):
  if i == 4 :
    continue
  print(i)
```

```python
i = 1
while i <= 10:
  if i == 4 :
    continue
  print(i)
  i += 1
# akan muncul hasil infinite loop, karena 2 perintah di bawah continue akan di skip, sehingga continue tidak boleh di atas update variabel
```
